```python
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA
import pandas as pd
import plotly.express as px
```

```python
wine = load_wine()
df = pd.DataFrame(wine.data, columns=wine.feature_names)
```

```python
df
```

```python
wine.target_names
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
for col in df.columns:
    df[col] = MinMaxScaler().fit_transform(df[col].values.reshape(-1, 1))
```

```python
df
```

```python
pca = PCA(n_components=3)

df_decompositioned = pd.DataFrame(pca.fit_transform(df))
df_decompositioned['label'] = wine.target.astype(str)
```

```python
px.scatter_3d(df_decompositioned, 0, 1, 2, color='label', height=800, width=800)
```

```python
from sklearn.decomposition import TruncatedSVD
```

```python
svd = TruncatedSVD(n_components=3)

df_decompositioned = pd.DataFrame(svd.fit_transform(df))
df_decompositioned['label'] = wine.target.astype(str)
```

```python
px.scatter_3d(df_decompositioned, 0, 1, 2, color='label', height=800, width=800)
```
