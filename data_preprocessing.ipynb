```python
import pandas as pd



df_games = pd.read_csv("datasets/archive/games.csv")
df_users = pd.read_csv("datasets/archive/users.csv")
# df_recommendations = pd.read_csv("datasets/archive/recommendations.csv")
```

```python
import json

tags = set()
objs = {}
with open("datasets/archive/games_metadata.json") as f:
    content = f.read()
    for i, obj in enumerate(content.split('\n')):
        try:
            d = json.loads(obj)
            objs[i] = d['tags']
            tags = tags | set(d['tags'])
        except:
            pass
tags
```

```python
objs
```

```python
s = pd.Series(objs)
df_tags = pd.get_dummies(s.explode()).groupby(level=0).sum()
```

```python
df_partial = pd.DataFrame(columns=['app_id', 'title'] + list(tags))
df_game_name = pd.read_csv('datasets/archive/games.csv')

df_full = pd.concat([df_game_name, df_tags], axis=1)
```
