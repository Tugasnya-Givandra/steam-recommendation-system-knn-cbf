```python
import pandas as pd



df_games = pd.read_csv("datasets/archive/games.csv")
df_users = pd.read_csv("datasets/archive/users.csv")
# df_recommendations = pd.read_csv("datasets/archive/recommendations.csv")
```

```python
import json

tags = set()
objs = {}
with open("datasets/archive/games_metadata.json", encoding="utf8") as f:
    content = f.read()
    for i, obj in enumerate(content.split('\n')):
        try:
            d = json.loads(obj)
            objs[d['app_id']] = d['tags']
            tags = tags | set(d['tags'])
        except:
            pass
tags = list(tags)
```

```python
df = pd.DataFrame(columns=['app_id'] + tags)
```

```python
d = {}
for i, (app_id, _tags) in enumerate(objs.items()):
    d[i] = {'app_id': app_id}
    d[i].update({tag:int(tag in _tags) for tag in tags})
    print(i)
```

```python
df_tags = pd.DataFrame.from_dict(d, "index")
```

```python
df_game_name = pd.read_csv('datasets/archive/games.csv')
```

```python
df_full = df_game_name.merge(df_tags, on=['app_id'])
```
